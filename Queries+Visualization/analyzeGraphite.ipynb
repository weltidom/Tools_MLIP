```python
import pandas as pd
import numpy as np
import plotly.express as px
import os
import py4vasp as pv
```

```python
pd.read_pickle('/Users/dominicwelti/Documents/Master_Thesis_Data_Set/Graphite/MACE/results.pkl')
```

```python
dft_folder='/Users/dominicwelti/Documents/Master_Thesis_Data_Set/Graphite/DFT'
files=list(filter(lambda x:'.h5' in x, os.listdir(dft_folder)))

data_h5=pd.DataFrame({
    'Calculation type': 'DFT',
    'Model': '',
    'c Factor': [0]*len(files),
    'c Parameter': 0,
    'Energy': 0
})
```

```python
for i, row in data_h5.iterrows():
    data_h5.loc[i, 'Model'] = files[i].split('_')[0]
    data_h5.loc[i, 'c Factor'] = files[i].split('_')[1].split('.h')[0]

    calc=pv.Calculation.from_file(f'{dft_folder}/{files[i]}')
    atoms=calc.structure.to_ase()
    data_h5.loc[i, 'c Parameter'] = atoms.get_cell()[2,2]
    data_h5.loc[i, 'Energy'] = calc.energy.read()['energy(sigma->0)']
```
