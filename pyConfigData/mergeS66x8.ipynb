```python
from pyDataHandler import *
```

```python
#retrieve paths
import os
folder_path='/Users/dominicwelti/Documents/Master_Thesis_Data_Set/s66x8'

files=[]
for (dirpath, dirnames, filenames) in os.walk(folder_path):
    for filename in filenames:
        files.append(f'{dirpath}/{filename}')

f=filter(lambda x: '.xyz' in x, files)
files=list(f)
```

```python
data_lst=[]

for file in files:
    subfolder=file.split('/')[-2]
    name=file.split('/')[-1]
    xyz=Xyz(f'{folder_path}/{subfolder}', name)
    xyz.read()
    data_lst.append(xyz.data)
```

```python
import pandas as pd

xyz.data=pd.concat(data_lst)
```

```python
xyz.path='/Users/dominicwelti/Documents/Master_Thesis_Data_Set/s66x8/Merged_dataset'
xyz.write(xyz.data)
```
